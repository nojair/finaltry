```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
X1 = pd.read_excel('motor-AG&EN.xlsx')
X2 = pd.read_excel('motor-AG&FT.xlsx')
X3 = pd.read_excel('motor-ZN.xlsx')
X4 = pd.read_excel('motor-PIDT.xlsx')
X5 = pd.read_excel('motor-PIDT.xlsx')
```

```python
large=21

outputData1 = list(X1['y'])
inputData1 = list(X1['u'])
inputMax1 = max(inputData1)
inputData1 = list(map(lambda x: x / inputMax1, inputData1[large:]))
outputData1 = list(map(lambda x: x / inputMax1, outputData1[large:]))

outputData2 = list(X2['y'])
inputData2 = list(X2['u'])
inputMax2 = max(inputData2)
inputData2 = list(map(lambda x: x / inputMax2, inputData2[large:]))
outputData2 = list(map(lambda x: x / inputMax2, outputData2[large:]))

outputData3 = list(X3['y'])
inputData3 = list(X3['u'])
inputMax3 = max(inputData3)
inputData3 = list(map(lambda x: x / inputMax3, inputData3[large:]))
outputData3 = list(map(lambda x: x / inputMax3, outputData3[large:]))

outputData4 = list(X4['y'])
inputData4 = list(X4['u'])
inputMax4 = max(inputData4)
inputData4 = list(map(lambda x: x / inputMax4, inputData4[large:]))
outputData4 = list(map(lambda x: x / inputMax4, outputData4[large:]))

outputData5 = list(X5['y'])
inputData5 = list(X5['u'])
inputMax5 = max(inputData5)
inputData5 = list(map(lambda x: x / inputMax5, inputData5[large:]))
outputData5 = list(map(lambda x: x / inputMax5, outputData5[large:]))
```

```python
def found_ts(vec):
    vec.reverse()
    n=0
    state=True
    for v in vec:
        if (v >= 0.97 and v <= 1.03 and state == True):
            n = n + 1
        else:
            state = False
    vec.reverse()
    print(n)
    return (229-n)*0.22848
```

```python
found_ts(outputData1)
```
