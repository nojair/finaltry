```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Fs = 1000 # Sampling frequency: 1000 Hz = 1 KHz       
T = 1/Fs                    
L = 1500
t = np.linspace(0,L-1,L)*T
```

```python
S = 5*np.sin(2*np.pi*50*t) #+ np.sin(2*np.pi*120*t);
X = S #+ 2*np.random.randn(len(t));
plt.plot(t[1:50], X[1:50])
plt.show()
```

```python
Y = np.fft.fft(X)
X_saved = np.fft.ifft(Y)

P2 = abs(Y/L)
P1 = P2[1:int(L/2+1)]
P1[2:] = 2*P1[2:]

f = Fs*np.arange(0,(L/2))/L
freq = np.fft.fftfreq(t.shape[-1])
# All spectre
plt.plot(freq[1400:1450],P2[1400:1450])
# Positive spectre
#plt.plot(freq[:int(L/2)],P1)

plt.show()
```
